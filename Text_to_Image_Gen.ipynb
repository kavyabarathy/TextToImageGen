```python
!pip install diffusers transformers accelerate torch torchvision torchaudio safetensors

import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import IPython.display as display

device = "cuda" if torch.cuda.is_available() else "cpu"
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16 if device == "cuda" else torch.float32)
pipe.to(device)

def generate_image(prompt):
    """Generate an image from text using Stable Diffusion."""
    print("Generating image, please wait...")
    image = pipe(prompt).images[0]
    return image


prompt = input("Enter an image generation prompt (Type cancel to exit): ")

while(prompt != "cancel"):
  generated_image = generate_image(prompt)

  display.display(generated_image)
  prompt = input("Enter another prompt (Type cancel to exit): ")

print("You have successfully exited this Image Generation Program")
```
